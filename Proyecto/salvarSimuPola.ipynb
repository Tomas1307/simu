```python
import numpy as np

# Parámetro de la distribución exponencial
lambda_inv = 0.009182652133589224

# Generamos 1000 datos que siguen una distribución exponencial
datos_exponenciales = np.random.exponential(1/lambda_inv, 19)

# Añadimos un valor aleatorio a cada dato para que no sea una perfecta exponencial
TEA_caja = datos_exponenciales + np.random.normal(0, 1, 19)  # Añade ruido gaussiano
TEA_caja
```

```python
newDFML = pd.DataFrame()

# Creamos la columna "llegada_fila" con longitud 23 y valor inicial NaN
newDFML["llegada_fila"] = [np.nan]*20

# Asignamos el primer valor
newDFML["llegada_fila"][0] = 45413

# Ahora podemos asignar los valores acumulativos
for i in range(1, len(newDFML)):
    if i != 0:
        newDFML["llegada_fila"][i] = newDFML["llegada_fila"][i-1] + TEA_caja[i]

newDFML
```

```python
import numpy as np

# Parámetros de la distribución uniforme
a = 10.0
b = 77.0

# Generamos 1000 datos que siguen una distribución uniforme
datos_uniformes = np.random.uniform(a, b, 20)

# Añadimos un valor aleatorio a cada dato
ts_caja_seg = datos_uniformes + np.random.normal(0, 1, 20)  # Añade ruido gaussiano

ts_caja_seg
```

```python
ts_caja_seg = [19.89722313, 11.47000604, 54.98955217, 60.99591531, 69.67517806,
       12.61776852, 12.97221808, 49.01950909, 16.71077862, 72.70757965,
       34.95076311, 31.52206058, 38.22991039, 31.81929656, 11.06006474,
       61.98621779, 52.6744726 , 70.4215763 , 34.74207554, 62.27521623]
```

```python
newDFML["inicio_servicio_caja"] = pd.Series([np.nan]*20)
newDFML["fin_servicio_caja"] = pd.Series([np.nan]*20)
```

```python
newDFML["ts_caja_seg"] = ts_caja_seg 
```

```python
newDFML["inicio_servicio_caja"][0] = 45459
```

```python
newDFML
```

```python
for i in range(0,len(newDFML["inicio_servicio_caja"])):
    for j in ts_caja_seg:
        if i == 0:
            newDFML["fin_servicio_caja"][i] = j +  newDFML["inicio_servicio_caja"][i]
        else:
            newDFML["inicio_servicio_caja"][i] = j +  newDFML["fin_servicio_caja"][i-1]
            newDFML["fin_servicio_caja"][i] = newDFML["inicio_servicio_caja"][i] + j
```

```python
newDFML["inicio_servicio_armado"] = pd.Series([np.nan]*20)
newDFML["fin_servicio_armado"] = pd.Series([np.nan]*20)
newDFML["ts_armado"] = pd.Series([np.nan]*20)
```

```python
newDFML['inicio_servicio_armado'][0] = newDFML['fin_servicio_caja'][0] + 65
```

```python
ts_armado = [93.39, 0.81, 170.07, 165.1, 102.29, 84.2, 69.52, 85.9, 
           179.55, 72.82, 117.2, 77.42, 86.33, 127.88, 126, 150.81,
           252.3, 259.44, 252.19, 110.48]
```

```python
newDFML['ts_armado'] = ts_armado
```

```python
for i in range(0,len(newDFML["inicio_servicio_armado"])):
    if i == 0:
        newDFML["fin_servicio_armado"][i] = newDFML["inicio_servicio_armado"][i] + newDFML["ts_armado"][i]

    else: 
        newDFML["inicio_servicio_armado"][i] = newDFML["fin_servicio_armado"][i-1] + newDFML["ts_armado"][i]
        newDFML["fin_servicio_armado"][i] = newDFML["inicio_servicio_armado"][i] + newDFML["ts_armado"][i]
```
