```python
import numpy as np

# Parámetro de la distribución exponencial
lambda_inv = 0.016269484644705443

# Generamos 1000 datos que siguen una distribución exponencial
datos_exponenciales = np.random.exponential(1/lambda_inv, 22)

# Añadimos un valor aleatorio a cada dato para que no sea una perfecta exponencial
TEA_caja = datos_exponenciales + np.random.normal(0, 0.8, 22)  # Añade ruido gaussiano
TEA_caja
```

```python
import numpy as np

# Parámetros de la distribución uniforme
a = 8.81
b = 85.83

# Generamos 1000 datos que siguen una distribución uniforme
datos_uniformes = np.random.uniform(a, b, 23)

# Añadimos un valor aleatorio a cada dato
ts_caja_seg = datos_uniformes + np.random.normal(0, 1, 23)  # Añade ruido gaussiano

ts_caja_seg
```

```python
newDFML["inicio_servicio_caja"] = pd.Series([np.nan]*23)
newDFML["fin_servicio_caja"] = pd.Series([np.nan]*23)
```

```python
newDFML["ts_caja_seg"] = ts_caja_seg 
```

```python
newDFML["inicio_servicio_caja"][0] = 45586
```

```python
for i in range(0,len(newDFML["inicio_servicio_caja"])):
    for j in ts_caja_seg:
        if i == 0:
            newDFML["fin_servicio_caja"][i] = j +  newDFML["inicio_servicio_caja"][i]
        else:
            newDFML["inicio_servicio_caja"][i] = j +  newDFML["fin_servicio_caja"][i-1]
            newDFML["fin_servicio_caja"][i] = newDFML["inicio_servicio_caja"][i] + j
```
